```python
import os
import pandas as pd

#import
school_data_csv = r"C:\Users\chelb\Documents\GitHub\pandas_challenge\PyCitySchools\schools_complete.csv"
student_data_csv = r"C:\Users\chelb\Documents\GitHub\pandas_challenge\PyCitySchools\students_complete.csv"

school_data = pd.read_csv(school_data_csv)
student_data = pd.read_csv(student_data_csv)

#merge
full_data = pd.merge(student_data, school_data, how="left", on=["school_name", "school_name"])
full_data.head()
```

```python
#District Summary

#Calculate the total number of schools
school_count = full_data["school_name"].nunique()
school_summ = {"Total Schools":[school_count]}
df_summary = pd.DataFrame(school_summ)

#Calculate the total number of students
student_count = full_data["Student ID"].nunique()
df_summary['Total Students'] = student_count

#Calculate the total budget
budget_list = full_data[['school_name','budget']].drop_duplicates()
budget_summ = budget_list['budget'].sum()
df_summary['Total Budget'] = budget_summ
#df_summary['Total Budget'] = df_summary['Total Budget'].map("${:,.2f}".format)

#Calculate the average math score
math_mean = full_data['math_score'].mean()
df_summary['Avg Math Score'] = math_mean

#Calculate the average reading score
read_mean = full_data['reading_score'].mean()
df_summary['Avg Reading Score'] = read_mean

#Calculate the percentage of students with a passing math score (70 or greater)
passing_math = full_data[full_data['math_score']>69]
pass_math_count = passing_math['Student ID'].nunique()
pass_math_mean = (pass_math_count/student_count)*100
df_summary['Passing Math'] = pass_math_mean

#Calculate the percentage of students with a passing reading score (70 or greater)
passing_read = full_data[full_data['reading_score']>69]
pass_read_count = passing_read['Student ID'].nunique()
pass_read_mean = (pass_read_count/student_count)*100
df_summary['Passing Reading'] = pass_read_mean
df_summary

#Calculate the overall passing rate (overall average score)
score_mean = (pass_math_mean + pass_read_mean)/2
df_summary['Overall Passing Score'] = score_mean

#format results
format_map = {'Total Students':'{:,.0f}','Total Budget':'${:,.2f}','Avg Math Score':'{:.2f}',
              'Avg Reading Score':'{:.2f}','Overall Passing Score':'{:.2f}%',
              'Passing Math':'{:.2f}%','Passing Reading':'{:.2f}%'}
for key, value in format_map.items():
    df_summary[key] = df_summary[key].apply(value.format)
df_summary
```

```python
#School Summary

#build school indexed dataframe with type, size, budget
s_df = full_data.drop_duplicates('School ID')
ss_df = s_df.set_index('school_name')
sch_df = pd.DataFrame(ss_df[['type','size','budget']])

    
#Total Students
#sch_stu_count = full_data.groupby('school_name')['Student ID'].count()
#pd.DataFrame(sch_stu_count)
#sch_df['Total Students'] = sch_stu_count
#sch_df

#per student budget
sch_df['Per Student Budget']=(sch_df['budget']/sch_df['size'])#.map("${:,.2f}".format) (cannot bin with $ format)

#Average Math Score
sch_math_mean = full_data.groupby('school_name')['math_score'].mean()
sch_df['Avg Math Score']= sch_math_mean#.map("{:.2f}".format)

#Average Reading Score
sch_read_mean = full_data.groupby('school_name')['reading_score'].mean()
sch_df['Avg Reading Score']= sch_math_mean#.map("{:.2f}".format)

# % Passing Math
sch_math_df = pd.DataFrame(full_data[['school_name','math_score']])
sch_pass_math_df = sch_math_df[sch_math_df['math_score'] > 69]
sch_pass_math_count = sch_pass_math_df.groupby('school_name')['math_score'].count()
sch_pass_math_mean = (sch_pass_math_count/sch_df['size'])*100
sch_df['% Passing Math'] = sch_pass_math_mean#.map("{:.2f}%".format)

# % Passing Reading
sch_read_df = pd.DataFrame(full_data[['school_name','reading_score']])
sch_pass_read_df = sch_read_df[sch_read_df['reading_score'] > 69]
sch_pass_read_count = sch_pass_read_df.groupby('school_name')['reading_score'].count()
sch_pass_read_mean = (sch_pass_read_count/sch_df['size'])*100
sch_df['% Passing Reading'] = sch_pass_math_mean#.map("{:.2f}%".format)

#Overall Passing Score
sch_score_mean = (sch_pass_math_mean + sch_pass_read_mean)/2
sch_df['% Overall Passing Score']= sch_score_mean#.map("{:.2f}%".format)

#format
format_map2 = {'size':'{:,.0f}','budget':'${:,.2f}'}
for key, value in format_map2.items():
    sch_df[key] = sch_df[key].apply(value.format)
sch_df
   
```

```python
#Math Scores by Grade
 
#grade df
ninth_df = full_data.loc[full_data['grade']=="9th",:]
tenth_df = full_data.loc[full_data['grade']=="10th",:]
eleventh_df = full_data.loc[full_data['grade']=="11th",:]
twelfth_df = full_data.loc[full_data['grade']=="12th",:]

#group df
ninth_group = ninth_df.groupby(['school_name'])
tenth_group = tenth_df.groupby(['school_name'])
eleventh_group = eleventh_df.groupby(['school_name'])
twelfth_group = twelfth_df.groupby(['school_name'])

#average df
ninth_mean = ninth_group['math_score'].mean()
tenth_mean = tenth_group['math_score'].mean()
eleventh_mean = eleventh_group['math_score'].mean()
twelfth_mean = twelfth_group['math_score'].mean()

math_grade_score = pd.DataFrame({"9th": ninth_mean,
                                "10th": tenth_mean,
                                "11th": eleventh_mean,
                                "12th": twelfth_mean})

format = {'9th':'{:.1f}', '10th':'{:.1f}', '11th':'{:.1f}', '12th':'{:.1f}'}
for key,value in format.items():
    math_grade_score[key]=math_grade_score[key].apply(value.format)
math_grade_score

```
